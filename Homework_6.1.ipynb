```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_excel("HW.xlsx")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.tree import export_text
```

```python
df = df[['Day','up/down','Where', 'Leader', 'Rain', 'Win']]
```

```python
fn=['up/down','Where', 'Leader', 'Rain', 'Win']
cn=['No', 'Yes']
df
```

```python
X = df[['up/down','Where', 'Leader', 'Rain']]
y = df[['Win']]
```

```python
clf = DecisionTreeClassifier(max_depth = 3,  random_state = 0)
clf.fit(X, y)
clf_pred = clf.predict(X)
clf_acc = accuracy_score(y, clf_pred)
print("%s accuracy %f" %  ("DecisionTreeClassifier" , clf_acc))
print(confusion_matrix(y, clf_pred))
print(classification_report(y, clf_pred))
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
plot_tree(clf, feature_names = fn, class_names=cn, filled = True);
```
